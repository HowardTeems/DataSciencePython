There are 4 features: sepal length, sepal width, petal length, and petal width.

Start by creating 2 clusters. Then standardize the data and try again. Does it make a difference?

Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
data = pd.read_csv('iris-dataset.csv')
data.head()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlim(0, 10)
plt.ylim(0, 7)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = data.iloc[:, 0:2]
x.head()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
iris_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters['cluster'] = iris_clusters
data_with_clusters
```

# Standardize the variables

Import and use the <i> scale </i> method from sklearn to standardize the data. 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
scaled_clusters = kmeans.fit_predict(x_scaled)
data_scaled_clusters = data.copy()
data_scaled_clusters['cluster'] = scaled_clusters
data_scaled_clusters
```

```python
plt.scatter(data_scaled_clusters['sepal_length'], data_scaled_clusters['sepal_width'], c=data_scaled_clusters['cluster'], cmap = 'rainbow')
plt.xlim(0, 10)
plt.ylim(0, 7)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []

for i in range(1, 9) :
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

wcss
```

### The Elbow Method

```python
number_clusters = range(1, 9)
plt.plot(number_clusters, wcss)
```

How many clusters are there?

Improvement is limited after 3 clusters.  2 is not bad either.
